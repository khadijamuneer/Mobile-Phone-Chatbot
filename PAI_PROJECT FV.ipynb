```python
import time
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

temp = {'Name': [], 'Review text': []}

driver.execute_script("window.scrollTo(0, 300);")
time.sleep(5)
driver.execute_script("window.scrollTo(300, 600);")
time.sleep(5)
driver.execute_script("window.scrollTo(600, 900);")
time.sleep(5)
driver.execute_script("window.scrollTo(900, 1100);")
time.sleep(5)
driver.execute_script("window.scrollTo(1100, 1300);")
time.sleep(5)
driver.execute_script("window.scrollTo(1300, 1600);")
time.sleep(5)


for page in range(0,3):
    try:
        reviews_container = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="module_product_review"]/div/div/div[3]/div[1]'))
        )
        reviews = reviews_container.find_elements(By.CLASS_NAME, 'item')

        for i, review in enumerate(reviews):
            try:
                content = review.find_element(By.CSS_SELECTOR, f"#module_product_review > div > div > div:nth-child(3) > div.mod-reviews > div:nth-child({i+1}) > div.item-content > div.content").text
                if content != '':
                    name = review.find_element(By.CLASS_NAME, 'middle').find_element(By.TAG_NAME, 'span').text[3:]
                    temp['Name'].append(name)
                    temp['Review text'].append(content)
            except NoSuchElementException as e:
                print(f'Error extracting review details: {e}')

        if page != 2:
            nxt_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="module_product_review"]/div/div/div[3]/div[2]/div/button[2]'))
            )
            driver.execute_script("arguments[0].click();", nxt_button)
    except NoSuchElementException:
        print('No More Pages')
        break

reviews = pd.DataFrame(temp)
print(reviews)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import random

                #this is the same code, i just filtered the brands by just splitting the first word
            
def scrape_product_info(driver):
    product_info_list = []

    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[1]/div/div[1]/div[2]/div[1]'))
    )

    product_elements = driver.find_elements(By.XPATH, '//*[@id="root"]/div/div[4]/div[1]/div/div[1]/div[2]/div')
    
    for product_element in product_elements:    
        name_xpath = './/div/div/div[2]/div[2]'
        price_xpath = './/div/div/div[2]/div[3]'

        name_element = WebDriverWait(product_element, 10).until(
            EC.visibility_of_element_located((By.XPATH, name_xpath))
        )
        price_element = WebDriverWait(product_element, 10).until(
            EC.visibility_of_element_located((By.XPATH, price_xpath))
        )

        name = name_element.text
        price = price_element.text

        brand = name.split()[0] 

        product_info = {
            'Name': name,
            'Price': price,
            'Brand': brand,
        }

        product_info_list.append(product_info)

    return product_info_list

def generate_random_ids(n):
    return [random.randint(1000, 9999) for _ in range(n)]

driver = webdriver.Chrome()  

final_out = []

for page_number in range(1, 6):
    driver.get(f"https://www.daraz.pk/smartphones/?page={page_number}")

    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div[4]/div[1]/div/div[1]/div[2]/div[1]'))
    )

    page_product_info = scrape_product_info(driver)

    final_out.extend(page_product_info)

# Generate random IDs for each product
random_ids = generate_random_ids(len(final_out))

# Create a separate dataset with IDs and Brands
id_dataset = pd.DataFrame({'ID': random_ids})
product_dataset = pd.DataFrame(final_out)
final_dataset = pd.concat([id_dataset, product_dataset], axis=1)

# Print the concatenated dataset
final_dataset


```

```python


# Save the concatenated dataset to a CSV file
final_dataset.to_csv('output_dataset.csv', index=False)

# Close the WebDriver
driver.quit()

print("Dataset saved to 'output_dataset.csv'")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_ratings(driver):
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//div[@class="c2prKC"]'))
    )

    product_elements = driver.find_elements(By.XPATH, '//div[@class="c2prKC"]')

    ratings_list = []
    for product_element in product_elements:
        rating_xpath = './/div[@class="c16H9d"]/div[2]/div[2]/div/div[2]/div/span/span'
        rating = WebDriverWait(product_element, 10).until(
            EC.visibility_of_element_located((By.XPATH, rating_xpath))
        ).text

        ratings_list.append({'Rating': rating})

    return ratings_list

driver = webdriver.Chrome()

final_out = []

for page_number in range(1, 6):
    driver.get(f"https://www.daraz.pk/smartphones/?page={page_number}")

    ratings = scrape_ratings(driver)
    final_out.extend(ratings)

df_ratings = pd.DataFrame(final_out)

# Append the ratings to your existing CSV
df_existing = pd.read_csv('output_dataset.csv')  # Replace with your actual file path
df_combined = pd.concat([df_existing, df_ratings], axis=1)

# Save the combined DataFrame to a new CSV
df_combined.to_csv('your_combined_file.csv', index=False)
```
